```python
import json

file_dir = "bcKG"

file_id2ent = f"{file_dir}/id2ent.json"
file_ent2id = f"{file_dir}/ent2id.json"
file_id2rel = f"{file_dir}/id2rel.json"
file_rel2id = f"{file_dir}/rel2id.json"

with open(file_id2ent, 'r') as file:
    kg_id2ent = json.load(file)
with open(file_ent2id, 'r') as file:
    kg_ent2id = json.load(file)
with open(file_id2rel, 'r') as file:
    kg_id2rel = json.load(file)
with open(file_rel2id, 'r') as file:
    kg_rel2id = json.load(file)

with open("../mnt/PubMedBERT/vocab.txt",'r') as f:
    lines = f.readlines()
    for line in lines:
        print(line)
    
```

```python
import numpy as np
import torch
from tqdm.auto import tqdm
from transformers import AutoTokenizer, AutoModel  


tokenizer = AutoTokenizer.from_pretrained('../mnt/PubMedBERT/')  
bert_model = AutoModel.from_pretrained('../mnt/PubMedBERT')
device = torch.device('cuda')
bert_model.to(device)
bert_model.eval()
```

```python
names = []
for idx in tqdm(range(len(kg_id2ent))):
    ent = kg_id2ent[str(idx)]  #id:ÂÆû‰Ωì
    names.append(ent)
print(names)
```

```python
embs = []
tensors = tokenizer(names, padding=True, truncation=True, return_tensors="pt")
with torch.no_grad():
    for i, j in enumerate(tqdm(names)):
        outputs = bert_model(input_ids=tensors["input_ids"][i:i+1].to(device), 
                               attention_mask=tensors['attention_mask'][i:i+1].to(device))
        out = np.array(outputs[1].squeeze().tolist()).reshape((1, -1))
        embs.append(out)
embs = np.concatenate(embs)
np.save("bcKG/ent_emb.npy", embs)
```

```python
import pickle
with open("bcKG/entity_embedding.pkl", "wb") as file:
    pickle.dump(embs, file)
```
